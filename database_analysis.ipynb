```python
import pandas as pd
import numpy as np
```

```python
#  my file name is 'diabetes_prediction_dataset.csv'
data = pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
data.head()
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Check for duplicate rows
print(data.duplicated().sum())
```

```python
# Remove duplicate rows
data.drop_duplicates(inplace=True)
```

```python
# Assuming your cleaned dataset name is 'cleaned_diabetes_dataset.csv'
data.to_csv('cleaned_diabetes_prediction_dataset.csv', index=False)
```

```python
# Load the cleaned dataset
cleaned_data = pd.read_csv('cleaned_diabetes_prediction_dataset.csv')
```

```python
# Number of rows and columns in the cleaned dataset
num_rows, num_cols = cleaned_data.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
# Count of male and female
gender_counts = cleaned_data['gender'].value_counts()
print("Male count:", gender_counts['Male'])
print("Female count:", gender_counts['Female'])
```

```python
# Average age of male and female
average_male_age = cleaned_data.loc[cleaned_data['gender'] == 'Male', 'age'].mean()
average_female_age = cleaned_data.loc[cleaned_data['gender'] == 'Female', 'age'].mean()
print("Average age of male:", average_male_age)
print("Average age of female:", average_female_age)
```

```python
# Average BMI
average_bmi = cleaned_data['bmi'].mean()
print("Average BMI:", average_bmi)
```

```python
# Average blood glucose level for male and female
average_male_glucose = cleaned_data.loc[cleaned_data['gender'] == 'Male', 'blood_glucose_level'].mean()
average_female_glucose = cleaned_data.loc[cleaned_data['gender'] == 'Female', 'blood_glucose_level'].mean()
print("Average blood glucose level for male:", average_male_glucose)
print("Average blood glucose level for female:", average_female_glucose)
```

```python
import matplotlib.pyplot as plt
```

```python
# Visualization 1: Gender Distribution Bar Chart
gender_counts = data['gender'].value_counts()
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'black'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Visualization 2: Age Distribution Histogram by Gender
plt.figure(figsize=(10, 6))
plt.hist(data[data['gender'] == 'Male']['age'], bins=20, alpha=0.5, label='Male', color='black')
plt.hist(data[data['gender'] == 'Female']['age'], bins=20, alpha=0.5, label='Female', color='blue')
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Visualization 3: Boxplot of BMI by Gender
plt.figure(figsize=(8, 6))
data.boxplot(column='bmi', by='gender')
plt.title('BMI Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('BMI')
plt.show()
```

```python
# Visualization 4: Blood Glucose Level Distribution Density Plot by Gender
plt.figure(figsize=(10, 6))
data[data['gender'] == 'Male']['blood_glucose_level'].plot(kind='density', label='Male', color='black')
data[data['gender'] == 'Female']['blood_glucose_level'].plot(kind='density', label='Female', color='red')
plt.title('Blood Glucose Level Distribution by Gender')
plt.xlabel('Blood Glucose Level')
plt.ylabel('Density')
plt.legend()
plt.show()
```
